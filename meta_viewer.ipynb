```python
aaa = 2

with open(path, "r+", encoding="utf8") as f:
    meta_reader = jsonlines.Reader(f)
    it = iter(meta_reader)
    for i in range(aaa):
        meta0 = next(it)
    for i in meta0:
        if i=='allProofs':
            continue
        if isinstance(meta0[i],dict):
            print(str(i)+':')
            for j in meta0[i]:
                if isinstance(meta0[i][j],dict):
                    print('\t'+str(j)+':')
                    for k in meta0[i][j]:
                        print(f'\t\t{k}:{meta0[i][j][k]}')
                else:
                    print(f'\t{j}:{meta0[i][j]}')
        else:
            print(f'{i}:{meta0[i]}')
    print('allProofs:')
    for i in meta0['allProofs'].split('@')[1:]:
        print('\t'+i)
    # print(meta0)
```
