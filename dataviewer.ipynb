```python
import jsonlines
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
def get_num_of_sentences_per_id_dist(filename):
    with open(filename, "r+", encoding="utf8") as f:
        contexts = []
        test_reader = jsonlines.Reader(f)
        for i in test_reader:
            contexts.append(i['context'].strip().split('.')[:-1])
        len_list = list(map(lambda x:len(x),contexts))
        len_dict = dict(zip(*np.unique(len_list,return_counts=True)))
        plt.bar(len_dict.keys(),len_dict.values())
        plt.title(f'{filename}\ntotal: '+str(len(contexts)))
        plt.show()
        # return len_dict
```

```python
get_num_of_sentences_per_id_dist('data/rule-reasoning-dataset-V2020.2.5.0/original/depth-0/train.jsonl')
get_num_of_sentences_per_id_dist('data/rule-reasoning-dataset-V2020.2.5.0/original/depth-1/train.jsonl')
get_num_of_sentences_per_id_dist('data/rule-reasoning-dataset-V2020.2.5.0/original/depth-2/train.jsonl')
get_num_of_sentences_per_id_dist('data/rule-reasoning-dataset-V2020.2.5.0/original/depth-3/train.jsonl')
get_num_of_sentences_per_id_dist('data/rule-reasoning-dataset-V2020.2.5.0/original/depth-5/train.jsonl')
get_num_of_sentences_per_id_dist('data/rule-reasoning-dataset-V2020.2.5.0/original/depth-3ext/train.jsonl')
```
